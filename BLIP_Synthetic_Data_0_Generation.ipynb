```python
from PIL import Image
import requests
from transformers import AutoProcessor, Blip2ForConditionalGeneration
import torch
from tqdm import tqdm
```

```python
processor = AutoProcessor.from_pretrained("Salesforce/blip2-opt-2.7b")
model = Blip2ForConditionalGeneration.from_pretrained("Salesforce/blip2-opt-2.7b", torch_dtype=torch.float16)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
```

```python
def predict_caption(image):
    inputs = processor(image, return_tensors="pt").to(device, torch.float16)

    outputs = model.generate(**inputs, max_new_tokens=20)

    generated_text = processor.batch_decode(outputs, skip_special_tokens=True)[0].strip()
    return generated_text
```

# Streaming a dataset
Streaming the dataset so it doesn't need to be downloaded.
Documentation: https://huggingface.co/docs/datasets/stream

```python
datastream = load_dataset("nlphuji/flickr30k", split="test", streaming=True).take(30000)

# to load test data, we need to use .skip() instead of .take()
# teststream = load_dataset("nlphuji/flickr30k", split="test", streaming=True).skip(30000)
```

```python
synthetic_schema = pa.schema([
  ('caption', pa.string()),
  ('split', pa.string()),
  ('img_id', pa.string()),
  ('filename', pa.string())
])
```

```python
synthetic_data_0 = []
counter = 0

filepath = '/content/drive/Shareddrives/Dudaa Drive/646 Vislang/synthetic_data_0/'

for item in tqdm(datastream):
  if counter%5000 == 0 and counter > 0:
    synthetic_data_table = pa.Table.from_pylist(synthetic_data_0)
    filename = 'synthetic_data_table_0_' + str(int(counter/5000)) + '.parquet'
    with open(filepath+filename, 'w') as f:
      pq.write_table(synthetic_data_table, f)
    synthetic_data_0 = []
    break

  image = item['image']
  predicted_caption = predict_caption(image)

  new_item = {
        'caption': [predicted_caption],
        'split': item['split'],
        'img_id': item['img_id'],
        'filename': item['filename']
    }

  synthetic_data_0.append(new_item)
  counter += 1
```
